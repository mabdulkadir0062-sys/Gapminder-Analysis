```python
import plotly.express as px
import polars as pl
from plotly.graph_objs.waterfall.increasing import marker

# Load dataset (initially Pandas via Plotly)
df_pd = px.data.gapminder()

# Convert to Polars
df = pl.from_pandas(df_pd)

# Aggregate per continent
df_continent = (
     df.group_by("continent", maintain_order=True)
     .agg([
      pl.mean("lifeExp", ).alias("avg_lifeExp"),
      pl.mean("gdpPercap").alias("avg_gdp"),
      pl.sum("pop").alias("total_pop")
])
).sort("total_pop", descending=True)

#convert to pandas for plotly
df_continent_pd = df_continent.to_pandas()

fig1 = px.bar(
    df_continent_pd,
    x="continent",
    y="avg_lifeExp",
    color="continent",
    title="Average Life Expectancy by Continent",
    hover_data=["total_pop", "avg_gdp"],

)
fig1.show()

fig2 = px.scatter(
    df_continent_pd,
    x="avg_gdp",
    y="avg_lifeExp",
    size="total_pop",
    color="continent",
    hover_name="continent",
    log_x=True,
    title="GDP vs Life Expectancy by Continent",
    trendline="ols"
)
fig2.show()

df_trend = (
    df.group_by(["continent","year"])
    .agg(
     pl.mean("lifeExp").alias("avg_lifeExp")
    )
    .sort("year", descending=True)
)

# convert to pandas for plotly
df_trend_pd = df_trend.to_pandas()

fig3 = px.line(
    df_trend_pd,
    x="year",
    y="avg_lifeExp",
    color="continent",
    title="Life Expectancy Trend by Continent",
    template="plotly_white"

)
fig3.show()
```
